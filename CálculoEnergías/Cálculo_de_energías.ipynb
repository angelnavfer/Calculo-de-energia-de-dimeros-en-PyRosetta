```python
!pip install pyrosettacolabsetup
import pyrosettacolabsetup
pyrosettacolabsetup.install_pyrosetta(cache_wheel_on_google_drive=False) # Set to False
import pyrosetta
from pyrosetta import *
from pyrosetta.rosetta import *
from pyrosetta.rosetta.core.scoring import ScoreType
from pyrosetta.rosetta.core.pose import Pose
from pyrosetta.rosetta.core.id import AtomID, TorsionID
from pyrosetta.teaching import *
import pyrosetta.toolbox
from pyrosetta import pose_from_sequence
from pyrosetta.toolbox import *
init()

# Iniciar PyRosetta
import pyrosetta
pyrosetta.init("-mute all")

# Cargar el documento .PDB de la proteína LuxR silvestre
print("Upload the wild-type PDB file:")
wt_upload = files.upload()
wt_filename = next(iter(wt_upload))

# Subir los documentos .PDB de los mutantes
print("Upload mutant PDB files:")
mutant_uploads = files.upload()

# Instalar la función para puntuar las estructuras
from pyrosetta import pose_from_pdb
scorefxn = pyrosetta.get_fa_scorefxn()

# Puntuar la variante silvestre
wt_pose = pose_from_pdb(wt_filename)
wt_score = scorefxn(wt_pose)
print(f"\nWT score: {wt_score:.2f}")

# Puntuar las variantes mutantes
results = [("WT", wt_score)]
for mutant_filename in mutant_uploads:
    pose = pose_from_pdb(mutant_filename)
    score = scorefxn(pose)
    label = os.path.splitext(mutant_filename)[0]
    results.append((label, score))

# Ordenar por energía
sorted_results = sorted(results, key=lambda x: x[1], reverse=True)

# Mostrarlo en una tabla
print("\nSorted scores (higher = worse), with relative energy (mutant / WT):")
for label, score in sorted_results:
    relative = score / wt_score
    print(f"{label}: Score = {score:.2f}, Relative to WT = {relative:.3f}")
```

```python
# Guardar los resultados en un archivo .CSV
import pandas as pd

# Generar un marco de datos
df = pd.DataFrame(sorted_results, columns=["Label", "Score"])
df["Relative_to_WT"] = df["Score"] / wt_score

# Guardarlo en un documento .CSV
csv_filename = "mutant_energy_comparison.csv"
df.to_csv(csv_filename, index=False)

# Descargar el documento .CSV
from google.colab import files
files.download(csv_filename)
```
